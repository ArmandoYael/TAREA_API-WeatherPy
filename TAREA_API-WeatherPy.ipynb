```python
#worldcities_pd["pressure"] = response_json[0]["list"][0]["main"]["pressure"]
#worldcities_pd.head()
identificador= []
nombre= []
nubes= []
pais = []
dt = []
humedad =[]
latitud =[]
longitud = []
temp_max = []
viento = []
presion = []

for x in range(len(response_json)):
    identificador.append(response_json[x]["list"][0]["id"])
    nombre.append(response_json[x]["list"][0]["name"])
    nubes.append(response_json[x]["list"][0]["clouds"]["all"])
    pais.append(response_json[x]["list"][0]["sys"]["country"])
    dt.append(response_json[x]["list"][0]["dt"])
    humedad.append(response_json[x]["list"][0]["main"]["humidity"])
    latitud.append(response_json[x]["list"][0]["coord"]["lat"])
    longitud.append(response_json[x]["list"][0]["coord"]["lon"])
    temp_max.append(response_json[x]["list"][0]["main"]["temp_max"])
    viento.append(response_json[x]["list"][0]["wind"]["speed"])
    presion.append(response_json[x]["list"][0]["main"]["pressure"])


TABLAFINAL = pd.DataFrame({"ID":identificador,
                           "CITY": nombre,
                           "CLOUDINESS": nubes,
                           "COUNTRY":pais,
                           "DATE":dt,
                           "HUMIDITY":humedad,
                           "LAT":latitud,
                           "LNG":longitud,
                           "MAX TEMP": temp_max,
                           "WIND SPEED":viento,
                           "PRESSURE":presion
                          })

TABLAFINAL.head()
```

```python
x = TABLAFINAL["LAT"]
y = TABLAFINAL["MAX TEMP"]

plt.title("CITY LATITUDE VS MAX TEMPERATURE")
plt.xlabel("LATITUDE")
plt.ylabel("MAX TEMPERATURE (K)")
plt.scatter(x,y,edgecolors="black",alpha=0.7)
plt.show()
```
